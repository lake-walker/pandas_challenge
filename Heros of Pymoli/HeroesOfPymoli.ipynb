```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Unique items
unique_items = purchase_data['Item Name'].unique()
unique_item_count = len(unique_items)
unique_item_count

# Average Purchase Price
total_price = purchase_data['Price'].sum()
avg_price = round(total_price / len(purchase_data['Price']), 2)
avg_price = '${:,.2f}'.format(avg_price)


# Total Number of Purchases
total_purchases = purchase_data['Purchase ID'].nunique()
total_purchases

# Total Revenue 
total_revenue = '${:,.2f}'.format(total_price)

# Create Dataframe for data
summary_df = pd.DataFrame({'Number of Unique Items': [unique_item_count],
                          'Average Purchase Price': [avg_price],
                          "Total Number of Purchases": [total_purchases],
                          'Total Revenue': [total_revenue]})
summary_df

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Count of Male Players
unique_players = purchase_data[['SN','Gender']].drop_duplicates()
males = unique_players[unique_players['Gender'] == 'Male']
male_count = males['Gender'].count()
male_count
# Count of Female Players
females = unique_players[unique_players['Gender'] == 'Female']
female_count = females['Gender'].count()
female_count
# Count of Other / Non-Disclosed
other = unique_players[unique_players['Gender'] == 'Other / Non-Disclosed']
#unique_players['Gender'].unique()
other_count = other['Gender'].count()
other_count
# Total Player Count
player_count = unique_players['SN'].count()
player_count

# Percentage of Males
percent_male = (male_count / player_count) * 100
percent_female = (female_count / player_count) * 100
percent_other = (other_count / player_count) * 100
# Format Percentages
percent_male = '{:,.2f}%'.format(percent_male)
percent_female = '{:,.2f}%'.format(percent_female)
percent_other = '{:,.2f}%'.format(percent_other)
percent_other
# Create DataFrame
gender_summary = pd.DataFrame({'Gender': ['Male', 'Female', 'Other / Non-Disclosed'],
                              'Total Count': [male_count, female_count, other_count],
                              'Percentage of Players': [percent_male, percent_female, percent_other]})
# Set Gender as Index
gender_summary = gender_summary.set_index('Gender')
gender_summary
```

```python

# Create bins for data and names
bins = [0, 9, 14, 19, 24, 29, 34, 39, 49]
age_ranges =  ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
age_df = purchase_data[['SN','Age','Gender','Price']]
# Categorize existing players
unique_players = age_df.drop_duplicates('SN')
pd.cut(age_df['Age'], bins, labels=age_ranges, include_lowest=True)
age_df
```
